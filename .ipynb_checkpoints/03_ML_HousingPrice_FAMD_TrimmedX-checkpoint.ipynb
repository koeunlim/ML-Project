### Factor Analysis for Mixed Data

```python
# Convert to mixed data
X_mixed = X_train_sel02.copy()
columns_categorical = ['GarageType_1.0','CentralAir_1.0','MSZoning_4.0','Condition1_2.0',
                       'GarageType_1.0','SaleType_6.0','Neighborhood_17.0']
for idx,col in enumerate(columns_categorical):
    X_mixed[col] = X_mixed[col].astype(object)
    
print(X_mixed.shape)
X_mixed.dtypes
X_cat = X_mixed.select_dtypes(["object","category"])
X_cont = X_mixed.select_dtypes(["float64","int64"]) 
print(X_cat.shape)
print(X_cont.shape)
```

```python
#Scale
scaler = StandardScaler()
X_mixed = pd.concat([X_cont,X_cat],axis=1)
X_mixed[X_cont.columns] = scaler.fit_transform(X_cont)
X_mixed.head()
```

```python
famd = prince.FAMD(
    n_components=28,
    n_iter=5,
    copy=True,
    check_input=True,
    engine='auto',
    random_state=2)
famd_X = famd.fit(X_mixed)
```

```python
FAMD_X = pd.DataFrame()
FAMD_X['ExplaineInertia'] = famd_X.explained_inertia_
FAMD_X['EigenValues'] = famd_X.eigenvalues_
FAMD_X['CummExplaineInertia'] = FAMD_X.ExplaineInertia.cumsum()
print(famd_X.total_inertia_)
FAMD_X
```

```python
famd_X_corr = famd_X.column_correlations(X_mixed).reset_index()
famd_X_corr
```

```python
famd_X_corr[famd_X_corr[0].abs() >= 0.5].sort_values(by = 0).plot(x='index',y=0,kind='barh')
```
