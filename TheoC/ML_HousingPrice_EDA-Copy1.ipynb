```python
# return a list with the names of columns that contain nans

cols_w_nan=[]
list_nan_obj = []

def list_nan_col(df):
    '''
    Creates a list of columns with any NA values
    '''
    for col in list(df):
        if np.sum(df[col].isnull()) > 0:
            cols_w_nan.append(col)
    return cols_w_nan


def null_in_obj(dataframe, list_of_columns):
    '''
    This function seeks out the null values in the columns 
    to replace them with the appropriate values based on 
    the data info
    '''
    for colname in list_of_columns:
        if dataframe[colname].dtype == 'object':
            list_nan_obj.append(colname)
    return list_nan_obj


# def fill_col_na(dataframe, list_of_columns):
#     '''
#     This function fills all NAs with 'None'
#     in the columns specified
#     '''
#     for colname in list_of_columns:
#         dataframe[colname] = dataframe[colname].fillna('None')
    

hp_dict_nan_col = {key: '' for key in list_nan_obj}
```
